```python
#가장 더웠던 날은?
import csv
f = open('/content/drive/MyDrive/Colab Notebooks/highSchool_Test/data/seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)
for row in data:
  if row[-1] == '':
    row[-1] = -999
  row[-1] = float(row[-1]) #최고 기온을 실수로 변환
  print(row)
f.close() 

#전처리)) 이상치값을 -999로 대체. 
# why? 우리나라 최고기온의 정상범위는 35~40도 정도. 이보다 현저히 낮은 값으로 넣어 처리하기로 함.
```

```python
#=======================================================final==========================================
#가장 더웠던 날은? 
import csv

f = open('/content/drive/MyDrive/Colab Notebooks/highSchool_Test/data/seoul.csv', 'r', encoding='cp949')

data = csv.reader(f)
header = next(data) #맨 윗줄을 header변수에 저장하기

max_temp = -999 #최고 기온 값을 저장할 변수
max_date = ''   #최고 기온이 가장 높았던 날을 저장할 변수

'''
만약 지금 까지의 최고 기온값보다 현재 행(row)의 최고 기온 값이 더 크다면 
   최고 기온 날짜 업데이트
   최고 기온 값 업데이트
'''
for row in data:
  if row[-1] == '':
    row[-1] = -999
  row[-1] = float(row[-1])

  if max_temp < row[-1]:
    max_date = row[0]
    max_temp = row[-1]

f.close() #파일닫기
print(max_date, max_temp)
print('기상 관측 이래 서울의 최고 기온이 가장 높았던 날은 ',max_date, "로, ", max_temp, '도 였습니다.')
```

#5. 데이터 시각화

> 데이터에 질문: 내 생일의 기온 변화를 그래프로 그리기



5.1 데이터 읽어오기

```python
import csv

f = open('/content/drive/MyDrive/Colab Notebooks/highSchool_Test/data/seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data) #헤더 부분을 next()함수를 사용해 제외시킴.

result = [] #최고 기온 데이터를 저장할 리스트 생성

for row in data:  
  if row[-1] != '':
    result.append(float(row[-1])) #result리스트에 최고 기온 값 추가

print(result)  
```

```python
#step2. 여름의 정점 8월의 최고기온 데이터만 추출해서 그래프 표현하기 => 날짜 데이터 추출하기

import csv
import matplotlib.pyplot as plt

f = open('/content/drive/MyDrive/Colab Notebooks/highSchool_Test/data/seoul2019.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data) #헤더 부분을 next()함수를 사용해 제외시킴.

result = [] #최고 기온 데이터를 저장할 리스트 생성

for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == '08': #8월에 해당하는 값이라면
      result.append(float(row[-1]))  #result리스트에 최고 기온값 추가

plt.plot(result, 'hotpink') #result리스트에 저장된 값을 hotpink색으로 그리기
plt.show()                  #그래프 그리기
```

##5.2 **데이터 시각화**...히스토그램

```python
#1월, 8월의 최고 기온 데이터만 추출해서 히스토그램 ...plt.hist()
import csv
import matplotlib.pyplot as plt

f = open('/content/drive/MyDrive/Colab Notebooks/highSchool_Test/data/seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data) #헤더부분을 netxt()함수를 사용해 제외시킴.

aug = [] #8월 최고기온 값을 저장할 리스트 aug생성
jan = [] #1월 최고기온 값을 저장할 리스트 jan생성

for row in data:
  if row[-1] != '' :
    if row[0].split('-')[1] == '08':
        aug.append(float(row[-1]))  
    if row[0].split('-')[1] == '01':
        jan.append(float(row[-1]))                   

plt.hist(aug, bins=100, color='r', label='Aug') #히스토그램
plt.hist(jan, bins=100, color='b', label='jan') #
plt.legend()

plt.show()
```

```python
#1월, 8월의 최고 기온 데이터만 추출해서 상자그림...plt.boxplot()
import csv
import matplotlib.pyplot as plt

f = open('/content/drive/MyDrive/Colab Notebooks/highSchool_Test/data/seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data) #헤더부분을 netxt()함수를 사용해 제외시킴.

aug = [] #8월 최고기온 값을 저장할 리스트 aug생성
jan = [] #1월 최고기온 값을 저장할 리스트 jan생성

for row in data:
  if row[-1] != '' :
    if row[0].split('-')[1] == '08':
        aug.append(float(row[-1]))  
    if row[0].split('-')[1] == '01':
        jan.append(float(row[-1]))                   

plt.boxplot([aug, jan]) #히스토그램
plt.show()
```

```python
#매달 최고 기온 데이터만 추출해서 상자그림...plt.boxplot()
import csv
import matplotlib.pyplot as plt

f = open('/content/drive/MyDrive/Colab Notebooks/highSchool_Test/data/seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data) #헤더부분을 netxt()함수를 사용해 제외시킴.

month = [ [],[],[],[],[],[],[],[],[],[],[],[] ] #12달, 월별 데이터를 저장할 month 생성(12개)

for row in data:
  if row[-1] != '' :
    #월과 같은 번호의 인덱스에 월별 데이터 저장( ex: 1월-> month[0] )
    month[int(row[0].split('-')[1])-1].append(float(row[-1])) #####

plt.boxplot(month) #박스플롯
plt.show()
```
